```python
image_set = torchvision.datasets.MNIST(root='../data/', train=True, download=True)
```

```python
im = np.array(image_set[0][0]).astype(np.float32)
plt.imshow(im)
```

```python
n = im.shape[0]
im_flat = im.flatten()

plt.figure()
plt.plot(im_flat)
```

```python
im_flat_stretched = change_duration(im_flat, n, 1, fmin=1, fmax=n//2)
plt.plot(im_flat_stretched)
```

```python
while np.sqrt(len(im_flat_stretched)) != int(np.sqrt(len(im_flat_stretched))): 
    im_flat_stretched = np.append(im_flat_stretched, 0)

n_prime = int(np.sqrt(len(im_flat_stretched)))
im_stretched = []

for i in range(n_prime):
    im_stretched.append(im_flat_stretched[i*n_prime:(i+1)*n_prime])
im_stretched = np.vstack(im_stretched)

plt.imshow(im_stretched)
```
